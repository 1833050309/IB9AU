**Key Insights**:
- Learned how to use `torch.tensor(..., requires_grad=True)` to track gradients.
- Used the `torch.sigmoid()` activation function inside a more complex expression.
- Understood how `.backward()` propagates gradients through the computational graph.
- Verified that gradients of multiple variables (w, x, b) are correctly computed and accessible via `.grad`.

---


```python
import torch
w = torch.tensor(2.0, requires_grad=True)
x = torch.tensor(4.0, requires_grad=True)
b = torch.tensor(1.5, requires_grad=True)
z = torch.sigmoid(w * x**2) + b**3
z.backward()
print("∂z/∂w:", w.grad)
print("∂z/∂x:", x.grad)
print("∂z/∂b:", b.grad)
```
