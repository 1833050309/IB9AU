**Key Insights**:

- Understood how `.backward()` propagates gradients through the computational graph.
- Verified that gradients of multiple variables (w, x, b) are correctly computed and accessible via `.grad`.

---


```python
import torch
w = torch.tensor(2.0, requires_grad=True)
x = torch.tensor(4.0, requires_grad=True)
b = torch.tensor(1.5, requires_grad=True)
z = torch.sigmoid(w * x**2) + b**3
z.backward()
print("∂z/∂w:", w.grad)
print("∂z/∂x:", x.grad)
print("∂z/∂b:", b.grad)
```

```python
import torch

torch.set_printoptions(precision=20, sci_mode=True)

w = torch.tensor(2.0, requires_grad=True, dtype=torch.float64)
x = torch.tensor(4.0, requires_grad=True, dtype=torch.float64)
b = torch.tensor(1.5, requires_grad=True, dtype=torch.float64)

z = torch.sigmoid(w * x**2) + b**3
z.backward()

print("z:", z.item())
print("dz/dw:", w.grad.item())
print("dz/dx:", x.grad.item())
print("dz/db:", b.grad.item())

u = w * x**2
s = torch.sigmoid(u)
ds_du = s * (1 - s)

dw_manual = (ds_du * x**2).item()
dx_manual = (ds_du * w * 2 * x).item()
db_manual = (3 * b**2).item()

print("manual dz/dw:", dw_manual)
print("manual dz/dx:", dx_manual)
print("manual dz/db:", db_manual)

print("abs error dw:", abs(w.grad.item() - dw_manual))
print("abs error dx:", abs(x.grad.item() - dx_manual))
print("abs error db:", abs(b.grad.item() - db_manual))
```
