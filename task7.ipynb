```python
import re
import textwrap
from urllib.request import Request, urlopen

URL = "https://www.fool.com/earnings/call-transcripts/2024/10/30/coinbase-global-coin-q3-2024-earnings-call-transcr/"

req = Request(
    URL,
    headers={
        "User-Agent": "Mozilla/5.0 (compatible; Task7Bot/1.0; +https://docs.python.org/3/)"
    },
)

html = urlopen(req).read().decode("utf-8", errors="ignore")
print("HTML length:", len(html))
print(html[:600])
```

```python
def strip_tags(html_text: str) -> str:

    html_text = re.sub(r"<script[\s\S]*?</script>", " ", html_text, flags=re.I)
    html_text = re.sub(r"<style[\s\S]*?</style>", " ", html_text, flags=re.I)

    text = re.sub(r"<[^>]+>", " ", html_text)

    text = re.sub(r"\s+", " ", text).strip()
    return text

raw_text = strip_tags(html)
print("Text length:", len(raw_text))
print(textwrap.fill(raw_text[:900], width=100))
```

```python
def find_snippets(text: str, patterns, window=220):
    hits = []
    for pat in patterns:
        for m in re.finditer(pat, text, flags=re.I):
            start = max(0, m.start() - window)
            end = min(len(text), m.end() + window)
            snippet = text[start:end]
            hits.append((pat, snippet))
    return hits

patterns = [
    r"diversify.*revenue",
    r"transaction revenue",
    r"subscription and services revenue",
    r"on pace to surpass \$?2 billion",
    r"total revenue was \$?1\.2 billion",
    r"total transaction revenue was \$?573 million",
    r"subscription and services revenue.*\$?556 million",
    r"trading volume was \$?185 billion",
    r"little-to-no fees on stablecoin",
    r"lower interest rates",
]

snips = find_snippets(raw_text, patterns)

print("Snippets found:", len(snips))
for i, (pat, snip) in enumerate(snips[:8], start=1):
    print(f"\n--- snippet {i} | pattern: {pat} ---")
    print(textwrap.fill(snip, width=110))
```

```python
def build_evidence_pack(snips, max_chars=4500):
    # 去重（按 snippet 文本）
    seen = set()
    cleaned = []
    for pat, snip in snips:
        s = snip.strip()
        if s in seen:
            continue
        seen.add(s)
        cleaned.append(f"[pattern={pat}] {s}")
    pack = "\n\n".join(cleaned)
    return pack[:max_chars]

evidence_pack = build_evidence_pack(snips)
print("Evidence pack chars:", len(evidence_pack))
print(textwrap.fill(evidence_pack[:900], width=110))
```

```python
import os, json
from openai import OpenAI

# 方式A：确保环境变量存在
assert os.environ.get("OPENAI_API_KEY"), "没找到 OPENAI_API_KEY。请先设置：os.environ['OPENAI_API_KEY']='...'"
client = OpenAI()

MODEL = "gpt-4o-mini"  # 先用兼容性最好的

def llm_answer(prompt: str, temperature: float = 0.2) -> str:
    resp = client.responses.create(
        model=MODEL,
        input=prompt,
        temperature=temperature,
    )
    return resp.output_text

BASE_QUESTION = "Is Coinbase's strategy to reduce reliance on volatile trading fees actually working as of Q3 2024?"

CONTEXT = f"""You are given excerpts from a Coinbase Q3 2024 earnings call transcript.
Use ONLY the evidence in the excerpts. If something is not in the excerpts, say you can't confirm.

EVIDENCE EXCERPTS:
{evidence_pack}
"""

prompt_A = f"""{CONTEXT}
Task: Answer the base question directly in 120-160 words.
- Start with a 1-sentence YES/NO leaning conclusion.
- Provide exactly 3 bullet evidence points, each quoting or paraphrasing a number from the excerpts.
- End with 1 sentence caveat.
Base question: {BASE_QUESTION}
"""

prompt_B = f"""{CONTEXT}
Role: You are an equity research analyst.
Write with headings:
1) Verdict
2) What changed in Q3 (numbers)
3) Why it matters for revenue volatility
4) Key remaining dependencies/risks (based on excerpts)
180-220 words.
Base question: {BASE_QUESTION}
"""

prompt_C = f"""{CONTEXT}
Approach: Metrics-first.
1) List key revenue mix numbers (total revenue, transaction revenue, subscription & services).
2) Explain whether the mix suggests reduced reliance on trading fees in Q3.
3) Mention at least one headwind from the excerpts that could still make revenues volatile.
160-210 words.
Base question: {BASE_QUESTION}
"""

prompt_D = f"""{CONTEXT}
Approach: Argue both sides.
- Para 1: strongest case it IS working (use numbers).
- Para 2: strongest case it is NOT fully working yet (use numbers/headwinds).
- Para 3: balanced conclusion (2-3 sentences).
190-230 words.
Base question: {BASE_QUESTION}
"""

ans_A = llm_answer(prompt_A)
ans_B = llm_answer(prompt_B)
ans_C = llm_answer(prompt_C)
ans_D = llm_answer(prompt_D)

print("===== Answer A =====\n", ans_A)
print("\n===== Answer B =====\n", ans_B)
print("\n===== Answer C =====\n", ans_C)
print("\n===== Answer D =====\n", ans_D)

judge_prompt = f"""
You are a strict judge. Evaluate four candidate answers to:
"{BASE_QUESTION}"

Score each answer 0-10 on:
1) Evidence grounding
2) Directly answers the question
3) Clarity/structure
4) Balanced nuance (mentions remaining volatility drivers)

Return JSON ONLY:
{{
  "scores": {{
    "A": {{"evidence":0,"direct":0,"clarity":0,"nuance":0,"total":0}},
    "B": {{"evidence":0,"direct":0,"clarity":0,"nuance":0,"total":0}},
    "C": {{"evidence":0,"direct":0,"clarity":0,"nuance":0,"total":0}},
    "D": {{"evidence":0,"direct":0,"clarity":0,"nuance":0,"total":0}}
  }},
  "winner": "A",
  "why_winner": "....",
  "improvements": {{"A":"...","B":"...","C":"...","D":"..."}}
}}

Answer A:
{ans_A}

Answer B:
{ans_B}

Answer C:
{ans_C}

Answer D:
{ans_D}
"""

judge_json = llm_answer(judge_prompt, temperature=0)
judge = json.loads(judge_json)

winner = judge["winner"]
best_answer = {"A": ans_A, "B": ans_B, "C": ans_C, "D": ans_D}[winner]
improve = judge["improvements"][winner]

final_prompt = f"""
Write an Executive Summary (~200 words, English) answering:
"{BASE_QUESTION}"

Must:
- Stay consistent with evidence excerpts.
- Incorporate strengths of the winning answer.
- Apply judge improvement note.
- Mention key Q3 mix numbers if present.
- End with 1 sentence implication for investors.

Winning answer ({winner}):
{best_answer}

Judge improvement note:
{improve}
"""

final_exec_summary = llm_answer(final_prompt)
print("\n===== FINAL Executive Summary (~200 words) =====\n", final_exec_summary)

print("\n===== Judge Result =====")
print(json.dumps(judge, indent=2, ensure_ascii=False))
```
