```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

df = pd.read_csv("credit_card_qa.csv")

X = df.drop("validity", axis=1)
y = df["validity"]

X = X.apply(lambda col: col.astype("category").cat.codes if col.dtypes == "object" else col)

print(y.value_counts())

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from tabpfn import TabPFNClassifier
from sklearn.metrics import accuracy_score
import torch

tabpfn_clf = TabPFNClassifier(device="cuda" if torch.cuda.is_available() else "cpu")
tabpfn_clf.fit(X_small, y_small)
y_pred_tabpfn = tabpfn_clf.predict(X_test_scaled)
acc = accuracy_score(y_test, y_pred_tabpfn)
print(f"{acc:.4f}")
```
